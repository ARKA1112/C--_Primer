```c++
//Ex9.4

bool foundnfound(vector<int>::iterator begin, vector<int>::iterator end, int value){
    for (vector<int>::iterator it = begin; it!= end; ++it){
        if(*it == value){
            return true;
        }
    }
    return false;
}

```

```c++
//Ex9.5

bool foundnfound(vector<int>::iterator begin, vector<int>::iterator end, int value){
    for (vector<int>::iterator it = begin; it!= end; ++it){
        if(*it == value){
            return &it;
        }
    }
    return false;
}
```
